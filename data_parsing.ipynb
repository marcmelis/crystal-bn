```python
import pandas as pd

dataframe = pd.read_csv("datasets/glass.data.txt")
dataframe.describe()
```

```python
interval = 20
for column in dataframe:
    if column != 'Type':
        cut = pd.cut(dataframe[column], interval)
        mapping_values = list(cut.sort_values().unique())
        parsed_data = [mapping_values.index(value) for value in cut]
        print (column)
        print (parsed_data)
```
